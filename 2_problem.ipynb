```python
import pandas as pd
import os
#file reading
olist_orders_dataset=pd.read_csv("/mnt/HC_Volume_18315164/home-jupyter/jupyter-f-denisenko-21/first_project/olist_orders_dataset.csv")
olist_orders_dataset
```

```python
no_deliveres=olist_orders_dataset.query('order_status!="delivered"')
# undelivered orders
no_deliveres
```

```python
#type of column with date was changed to form
#where there is month and year but day is 01 always
no_deliveres.order_estimated_delivery_date=no_deliveres.order_estimated_delivery_date.astype('datetime64[M]')
no_deliveres
```

```python
#Firstly, let's find the amount of every undelivery reason
# then we need to find mean value for every reason per month
no_deliveres=no_deliveres.groupby(['order_estimated_delivery_date',\
                                   'order_status'],as_index=False)\
    .agg({'order_id':'count'})\
    .groupby('order_status',as_index=False)\
    .agg({'order_id':'mean'})\
    .rename(columns={'order_id':'mean'})
#values was sorted for comfortabl reading
no_deliveres=no_deliveres.sort_values('mean')
no_deliveres
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(9,5))
#we change the size of plot window
sns.barplot(x='order_status',y='mean',data=no_deliveres)
#and we creat a bar plot
```
