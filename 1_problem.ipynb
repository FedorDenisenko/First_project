```python
#file reading
olist_customers_dataset=pd.read_csv("/mnt/HC_Volume_18315164/home-jupyter/jupyter-f-denisenko-21/first_project/olist_customers_dataset.csv")
olist_customers_dataset.nunique()
```

```python
customers_and_orders=olist_customers_dataset.merge(olist_orders_dataset,on='customer_id')
customers_and_orders
# merge customers and order information in one table
```

```python
import numpy as np
Search_NAN=pd.isna(customers_and_orders.order_approved_at)
NaN_in_order_approved_at=pd.DataFrame({'NaN_in':Search_NAN,'customer_id':customers_and_orders.customer_id})
NaN_in_order_approved_at
#table was created with flag of payment happend or not
#in the column "NaN_in" value "False" shows payment happend
```

```python
orders_and_NaN_approved=customers_and_orders.merge(NaN_in_order_approved_at, on='customer_id')
orders_purchase=orders_and_NaN_approved.query('NaN_in!=True')
orders_purchase
#Onnly rows was choosen which had became purchases
#Finally, the table "orders_purchase" has only purchases
```

```python
#Let's find the amount of users, which has done only one purchase
count_of_purchase=orders_purchase.groupby('customer_unique_id', as_index=False).agg({'order_id':'count'})
count_of_purchase.rename(columns={'order_id':'count_of_orders'})
```

```python
count_person_of_one_purchase=count_of_purchase.query('order_id==1')
count_person_of_one_purchase.shape
#Answer is 93049 of users, which has done only one purchase
```
