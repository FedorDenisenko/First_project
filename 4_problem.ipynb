```python
#Files reading
olist_customers_dataset=pd.read_csv("/mnt/HC_Volume_18315164/home-jupyter/jupyter-f-denisenko-21/first_project/olist_customers_dataset.csv")
olist_customers_dataset
```

```python
customers_and_orders=olist_customers_dataset.merge(olist_orders_dataset,on='customer_id')
customers_and_orders
#Merging customers and order information into one table
```

```python
import numpy as np
Search_NAN=pd.isna(customers_and_orders.order_approved_at)
customers_and_orders['NaN_in']=Search_NAN
only_purchers=customers_and_orders.query('NaN_in==False')
only_purchers
#table was created with flag of payment happend or not
#in the column "NaN_in" value "False" shows payment happend
#Table "only_purchers" has only purchases
```

```python
#creating the table with the type of data and the number of month
only_purchers['Month']=only_purchers.order_approved_at
only_purchers['Month']=only_purchers.Month.apply(pd.to_datetime)
only_purchers['Month']=only_purchers.Month.dt.strftime('%m')
only_purchers
```

```python
# creating a column "Year" with type of date and number of year
# to not be confused in what year every month is
only_purchers['Year']=only_purchers.order_approved_at
only_purchers['Year']=only_purchers.Year.apply(pd.to_datetime)
only_purchers['Year']=only_purchers.Year.dt.strftime('%Y')
only_purchers
```

```python
#Finding quantity of purchases for every user for every month
count_of_purches=only_purchers.\
    groupby(['customer_unique_id', 'Month','Year'],as_index=False)\
    .agg({'order_id':'count'})
count_of_purches
```

```python
#It is a function for definition of the coefficient of the number of weeks in a month
def find_count_of_week_in_month(x):
    if x=='01' or x=='03' or x=='05' or x=='07' or x=='08' or x=='10' or\
    x=='12':
        return 31/7
    elif x=='04' or x=='06' or x=='09' or x=='11':
        return 30/7
#    elif x==2 and (y==2004 or y==2008 or y==2012 or y==2016\
#    or y==2020 or y==2024):
#        return 29/7
    else:
        return 28/7
    
count_of_purches['count_of_week']=count_of_purches['Month']\
.apply(find_count_of_week_in_month)
#count_of_purches
count_of_purches
```

```python
#Putting in column "mean" the mean amount of purchases for every user
#per meek for every month
count_of_purches['mean']=count_of_purches['order_id']\
    /count_of_purches['count_of_week']
count_of_purches=count_of_purches.sort_values('mean',ascending=False)
count_of_purches.head(20)
```
