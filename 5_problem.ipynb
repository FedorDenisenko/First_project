```python
#Files reading
olist_customers_dataset=pd.read_csv("/mnt/HC_Volume_18315164/home-jupyter/jupyter-f-denisenko-21/first_project/olist_customers_dataset.csv")
olist_customers_dataset
```

```python
customers_and_orders=olist_customers_dataset.merge(olist_orders_dataset,on='customer_id')
customers_and_orders
#mearging customers and order information into one table
```

```python
import numpy as np
Search_NAN=pd.isna(customers_and_orders.order_approved_at)
customers_and_orders['NaN_in']=Search_NAN
only_purchers=customers_and_orders.query('NaN_in==False')
only_purchers
#table was created with flag of payment happend or not
#in the column "NaN_in" value "False" shows payment happend
#Table "only_purchers" has only purchases
```

```python
#Let's start to make a dataframe "kogorts"
#We creat column "Date_of_purches" by copying
# the column "order_approved_at" from "only_purchers"
#And we change the data type leaving only months
kogorts=pd.DataFrame()
kogorts['Date_of_purches']=only_purchers.order_approved_at
kogorts['Date_of_purches']=kogorts.Date_of_purches.apply(pd.to_datetime)
kogorts['Date_of_purches']=kogorts.Date_of_purches.\
    to_numpy().astype('datetime64[M]')
kogorts
```

```python
#adding column of unique users Â«id_customerÂ», Â«order_idÂ» 
#and Â«cohortÂ» like a minimum date
from pandas.tseries.offsets import DateOffset
kogorts['id_customer']=only_purchers.customer_unique_id
kogorts['order_id']=only_purchers.order_id
kogorts=kogorts.drop_duplicates()
kogorts['cohort']=kogorts\
    .groupby('id_customer')['Date_of_purches'].transform('min') 
kogorts
```

```python
kogorts_new=kogorts\
    .groupby(['cohort', 'Date_of_purches'], as_index=False)\
    .agg(user_count=('id_customer', 'nunique'))
kogorts_new
#finding the quantity of users in each group
```

```python
#creating a column Â«period_numberÂ» for show
#the duration between the first purchase and the next
kogorts_new['period_number'] = (kogorts_new.Date_of_purches\
    - kogorts_new.cohort)
kogorts_new.period_number=kogorts_new.period_number/np.timedelta64(1, 'M')
kogorts_new.period_number=kogorts_new.period_number.round(0)
kogorts_new.period_number= kogorts_new.period_number.convert_dtypes()
kogorts_new
```

```python
# Creating a pivot-table
kogorts_pivot = kogorts_new.pivot_table (index='cohort', 
                                     columns='period_number', 
                                     values='user_count')
kogorts_pivot
```

```python
#show the values of first row, which are the size of cohort
cohort_size=kogorts_pivot.iloc[:,0]
#getting a percantage
retention_matrix=kogorts_pivot.divide(cohort_size, axis=0)
retention_matrix
```

```python
#editing of visualisation
import matplotlib.pyplot as plt
import seaborn as sns
fig, ax = plt.subplots(1, 2, figsize=(12, 8),\
    sharey=True,gridspec_kw={'width_ratios': [1, 11]})
sns.heatmap(retention_matrix, 
            mask=retention_matrix.isnull(), 
            annot=True, 
            fmt='.0%', 
            cmap='RdYlGn', 
            ax=ax[1])
ax[1].set_title('Monthly storage', fontsize=16)
ax[1].set(xlabel='â„– periods', ylabel='')
cohort_size_df = pd.DataFrame(cohort_size)\
                             .rename(columns={0: 'cohort size'})

sns.heatmap(cohort_size_df, 
            annot=True, 
            cbar=False, 
            fmt='g', 
            ax=ax[0])
fig.tight_layout()
```
