```python
import pandas as pd
import os
#file reading
olist_orders_dataset=pd.read_csv("/mnt/HC_Volume_18315164/home-jupyter/jupyter-f-denisenko-21/first_project/olist_orders_dataset.csv")
olist_orders_dataset
```

```python
#file reading
olist_order_items_dataset=pd.read_csv("/mnt/HC_Volume_18315164/home-jupyter/jupyter-f-denisenko-21/first_project/olist_order_items_dataset.csv")
olist_order_items_dataset
```

```python
items_and_orders=olist_order_items_dataset.merge(olist_orders_dataset,on='order_id')
items_and_orders
#mearging customers and order information into one table
```

```python
import numpy as np
Search_NAN=pd.isna(items_and_orders.order_approved_at)
items_and_orders['NaN_in']=Search_NAN
only_purchers=items_and_orders.query('NaN_in==False')
only_purchers
#table was created with flag of payment happend or not
#in the column "NaN_in" value "False" shows payment happend
#Table "only_purchers" has only purchases
```

```python
# changing tha type of column 'order_approved_at'  
# and making from it the day of the week
only_purchers.order_approved_at=only_purchers.order_approved_at.apply(pd.to_datetime)
only_purchers.order_approved_at=only_purchers.order_approved_at.dt.strftime('%A')
only_purchers
```

```python
#Finding how ,amy times every product was included into purchase
day_of_week_and_product=only_purchers\
    .groupby(['product_id','order_approved_at'],as_index=False)\
    .agg({'order_id':'count'})
# renaming the column
day_of_week_and_product=day_of_week_and_product\
    .rename(columns={'order_id':'count_of_day'})
day_of_week_and_product
```

```python
#finding the max purchase quantity per day for every product
max_day_of_week_and_product=day_of_week_and_product\
    .groupby('product_id',as_index=False)\
    .agg({'count_of_day':'max'})
max_day_of_week_and_product
```

```python
# Getting for every product the day of the week, when
# the product is bought the most often
total=max_day_of_week_and_product.merge(day_of_week_and_product,\
                                        on=['product_id','count_of_day'])
total=total.sort_values('count_of_day', ascending=False)
total.head(20)
```
